```python
# Cell 2: Kịch bản Cài đặt (Đã sửa lỗi)

# 1. Thêm GPG Key chính thức của MongoDB
print("Đang thêm GPG key của MongoDB...")
!wget -qO- https://www.mongodb.org/static/pgp/server-7.0.asc | gpg --dearmor > /etc/apt/trusted.gpg.d/mongodb-server-7.0.gpg

# 2. Thêm kho lưu trữ MongoDB vào sources list
print("Đang thêm MongoDB vào sources list...")
!echo "deb [ arch=amd64,arm64 ] https://repo.mongodb.org/apt/ubuntu jammy/mongodb-org/7.0 multiverse" | tee /etc/apt/sources.list.d/mongodb-org-7.0.list > /dev/null

# 3. Cập nhật 'apt' để nhận kho lưu trữ mới
print("Đang cập nhật apt...")
!apt-get update > /dev/null

# 4. Cài đặt gói 'mongodb-org'
print("Đang cài đặt MongoDB Server, Shell, và Router (mongodb-org)...")
!apt-get install -y mongodb-org > /dev/null
print("Cài đặt hoàn tất!")

# 5. Tạo tất cả các thư mục dữ liệu
print("Đang tạo các thư mục dữ liệu...")
!mkdir -p /data/rs0-1 /data/rs0-2 /data/rs0-3
!mkdir -p /data/config1 /data/config2 /data/config3

# --- PHẦN 1: REPLICATION (Dùng làm Shard 1) ---
print("\n--- Bắt đầu PHẦN 1: REPLICATION ---")

# 6. Khởi động 3 node 'myReplicaSet' (ports 27017, 27018, 27019)
# ===== SỬA LỖI: Thêm cờ --shardsvr =====
print("Đang khởi động 3 node cho 'myReplicaSet' (với cờ --shardsvr)...")
!mongod --shardsvr --port 27017 --dbpath /data/rs0-1 --replSet "myReplicaSet" --logpath /data/rs0-1.log --fork
!mongod --shardsvr --port 27018 --dbpath /data/rs0-2 --replSet "myReplicaSet" --logpath /data/rs0-2.log --fork
!mongod --shardsvr --port 27019 --dbpath /data/rs0-3 --replSet "myReplicaSet" --logpath /data/rs0-3.log --fork
# ========================================

# 7. Khởi tạo Replica Set
print("Đang khởi tạo 'myReplicaSet'...")
!sleep 5 # Đợi 5 giây cho server khởi động
!mongosh --port 27017 --eval 'rs.initiate({ _id: "myReplicaSet", members: [ { _id: 0, host: "localhost:27017" }, { _id: 1, host: "localhost:27018" }, { _id: 2, host: "localhost:27019" } ] })'

# --- PHẦN 2: SHARDING ---
print("\n--- Bắt đầu PHẦN 2: SHARDING ---")

# 8. Khởi động 3 Config Server (ports 27050, 27051, 27052)
print("Đang khởi động 3 node Config Server 'configReplSet'...")
!mongod --configsvr --port 27050 --dbpath /data/config1 --replSet "configReplSet" --logpath /data/config1.log --fork
!mongod --configsvr --port 27051 --dbpath /data/config2 --replSet "configReplSet" --logpath /data/config2.log --fork
!mongod --configsvr --port 27052 --dbpath /data/config3 --replSet "configReplSet" --logpath /data/config3.log --fork

# 9. Khởi tạo Config Server Replica Set
print("Đang khởi tạo 'configReplSet'...")
!sleep 5 # Đợi 5 giây
!mongosh --port 27050 --eval 'rs.initiate({ _id: "configReplSet", configsvr: true, members: [ { _id: 0, host: "localhost:27050" }, { _id: 1, host: "localhost:27051" }, { _id: 2, host: "localhost:27052" } ] })'

# 10. Khởi động Query Router (mongos) (port 27060)
print("Đang khởi động Query Router (mongos)...")
!mongos --port 27060 --configdb "configReplSet/localhost:27050,localhost:27051,localhost:27052" --logpath /data/mongos.log --fork

# 11. Cấu hình Sharded Cluster (qua mongos)
print("Đang cấu hình Sharded Cluster (add shard, enable sharding)...")
!sleep 5 # Đợi 5 giây cho mongos khởi động

# Thêm 'myReplicaSet' làm Shard đầu tiên
print("BƯỚC 11a: Đang thêm Shard 'myReplicaSet'...")
!mongosh --port 27060 --eval 'sh.addShard("myReplicaSet/localhost:27017")'

# Thêm 10 giây chờ
print("BƯỚC 11b: Đang đợi Shard được đăng ký (10 giây)...")
!sleep 10

# Bật sharding cho database 'salesdb'
print("BƯỚC 11c: Đang bật sharding cho 'salesdb'...")
!mongosh --port 27060 --eval 'sh.enableSharding("salesdb")'

# Phân mảnh collection 'salesdb.orders'
print("BƯSỚC 11d: Đang phân mảnh collection 'salesdb.orders'...")
!mongosh --port 27060 --eval 'sh.shardCollection("salesdb.orders", { "customerId": 1 } )'

print("\n--- HOÀN THÀNH CÀI ĐẶT TOÀN BỘ CLUSTER ---")

# 12. Kiểm tra trạng thái cuối cùng
print("\n--- Trạng thái Replica Set (myReplicaSet) ---")
!mongosh --port 27017 --eval 'rs.status()' | grep -E "name|stateStr"

print("\n--- Trạng thái Sharded Cluster (mongos) ---")
!mongosh --port 27060 --eval 'sh.status()'
```
